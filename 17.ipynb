```python
DF = pd.read_csv("data//DF.csv")
```

```python
DF_F = DF[["urlAT","categoríaF"]].dropna()["urlAT"].to_list()
DF_LN = DF[["urlAT","categoríaLN"]].dropna()["urlAT"].to_list()
```

```python
DF_F
```

```python
DF_LN
```

```python
L = set(DF_F + DF_LN)
L
```

```python
L = [l.replace('obra','opiniones') for l in L]
L = [l+'?pagina=' for l in L]
L
```

```python
len(L)
```

```python
def getOpiniones(i, l = L):
    url  = L[i]
    m = AT.opinionesObra_getOpiniones(url,AT.opinionesObra_pagMax(url))
    return m
```

```python
def run_parallel(process_func, range_args):
    with concurrent.futures.ThreadPoolExecutor() as executor:
        futures = [executor.submit(process_func, i) for i in range_args]
        results = [f.result() for f in concurrent.futures.as_completed(futures)]
    return results
```

```python
r = run_parallel(getOpiniones, range(0,317))
```

```python
r = [i for i in r if i!=[]]
R = []
for i in r:
    if type(i)==str: R.append(i)
    if type(i)==list: R.extend(i)
```

```python
R = [i[0] for i in R if i!=[]]
R = ["https:"+i for i in R]
R = [i+"?pagina=" for i in R]
```
