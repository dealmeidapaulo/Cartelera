```python
import pandas as pd
import numpy as np
from lib.lib_threads import run_parallel
```

```python
DF = pd.read_csv("data/DF.csv")
```

```python
DF.head()
```

```python
for i in range(0,16676):
    try:
        with open('data/raw/{int_archivo}.txt'.format(int_archivo = i), 'r') as f:
            data = eval(f.read())
            for j in data[2]:
                if j[1:] not in DF.columns.to_list(): DF[j[1:]] = np.nan
                DF.loc[DF["urlAT"] == data[0],DF[j[1:]]] = 1
            for j in data[1]:
                j = j.replace("https://www.alternativateatral.com//","")
                if j not in DF.columns.to_list(): DF[j] = np.nan
                DF.loc[DF["urlAT"] == data[0],DF[j]] = 1

    except:
            print(i)
```

```python
DF.to_csv("data/DF_.csv", encoding='utf-8', index=False)
```
