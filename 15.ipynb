```python
import pandas as pd
```

VOY A QUERER LIMPIAR LOS DATOS PARA PODER HACER REGRESIONES.

```python
df_x1 = pd.read_csv("data//df_x1.csv")
df_x3 = pd.read_csv("data//df_x3.csv")
df_x4 = pd.read_csv("data//df_x4.csv")
```

```python
df_x1 = df_x1[["urlAT","categoríaF"]]
df_x1
```

```python
df_x3 = df_x3[["urlAT","categoríaLN"]]
df_x3 = df_x3.dropna()
df_x3
```

```python
df_x4 = df_x4[["urlAT","categoríaAT"]]
df_x4 = df_x4.dropna()
df_x4 = df_x4.loc[df_x4["categoríaAT"]>0]
df_x4 = df_x4[~df_x4["urlAT"].str.contains("opinador")]
df_x4
```

EXTRAIGO Y DOY FORMATO A LA INFORMACIÓN DE df_x4.
CREO MI DATAFRAME Y AGREGO LA INFORMACIÓN DE df_x4.

```python
dict_volumenAT = df_x4["urlAT"].value_counts().to_dict()
```

```python
list_urlAT = list(set(df_x4["urlAT"].values))
```

```python
dict_categoríaAT = df_x4.groupby(['urlAT'])['categoríaAT'].mean().to_dict()
```

```python
dataAT = {"urlAT":list_urlAT,"categoríaAT":[dict_categoríaAT[url] for url in list_urlAT],"volumenAT":[dict_volumenAT[url] for url in list_urlAT]}
```

```python
DF = pd.DataFrame(columns=["urlAT","categoríaAT","categoríaF","categoríaLN","volumenAT"], data = dataAT)
DF
```

EXTRAIGO Y DOY FORMATO A LA INFORMACIÓN DE df_x1.
AGREGO LA INFORMACIÓN DE df_x1 A MI DATAFRAME.

```python
dict_categoríaF = df_x1.groupby(['urlAT'])['categoríaF'].mean().to_dict()
```

```python
for key in dict_categoríaF.keys():
    DF.loc[DF["urlAT"] == key,"categoríaF"] = dict_categoríaF[key]
```

```python
DF
```

EXTRAIGO Y DOY FORMATO A LA INFORMACIÓN DE df_x3.
AGREGO LA INFORMACIÓN DE df_x3 A MI DATAFRAME.

```python
dict_categoríaLN = df_x3.groupby(['urlAT'])['categoríaLN'].mean().to_dict()
```

```python
for key in dict_categoríaLN.keys():
    DF.loc[DF["urlAT"] == key,"categoríaLN"] = dict_categoríaLN[key]
```

```python
DF
```

GUARDO MI DATAFRAME.

```python
DF.to_csv("data/DF.csv", encoding='utf-8', index=False)
```
