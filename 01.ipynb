Voy a querer obtener la información disponible de todas las obras publicadas en Farsa Mag.

https://farsamag.com/


```python
import lib.lib_soup_farsamag as farsamag
import lib.lib_soup as sp

import numpy as np
import pandas as pd
```

```python
#Consigo url específico de cada obra.
X = farsamag.Obras_getUrls()
```

```python
#Me quedo con el soup de cada una.
Y = []
for i in X:
    print(X.index(i))
    try:
       Y.append(sp.url_to_soup(i))
    except:
        Y.append("----   "+i)
```

```python
#Obtengo (url, título, categoría) de cada obra.
if len(X) == len(Y):
    X = [(X[Y.index(y)],farsamag.obra_getTitulo(y), farsamag.obra_getCategoria(y)) for y in Y]
```

```python
X
```

```python
X = [x for x in X if x[2] in ("Reseña","Imperdible","Destacada")] + [(x[0],x[1],"Desconocido") for x in X if not x[2] in ("Reseña","Imperdible","Destacada")]
```

```python
X = np.array(X)
```

```python
df = pd.DataFrame(X,columns=["url","nombre","categoría"])
```

```python
df.to_csv("data/obras_cola.csv", encoding='utf-8', index=False)
```
